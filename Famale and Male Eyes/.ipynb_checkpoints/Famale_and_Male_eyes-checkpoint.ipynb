```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2 as cv
import os
from PIL import Image
import re

import warnings
warnings.filterwarnings("ignore")
```

### Renamed Images

```python
def image_rename(path, name):
    i = 1
    for filenames in os.listdir(path):
        try:
            f, extension = os.path.splitext(filenames)
            names = path + filenames
            new_names = path + name + str(i) + extension
            os.renames(names, new_names)
            i = i + 1
        except Exception as e:
            print(e)
            i = i + 1
            continue
famale_path = "femaleeyes//"
male_path = "maleeyes//"

famale_lines = "famales"
male_lines = "males"

image_rename(famale_path, famale_lines) 
image_rename(male_path, male_lines) 
```

### Resized Images

```python
def resize_images(path):
    for filename in os.listdir(path):
        try:
            img = cv.imread(path + filename)
            img = cv.resize(img, (64, 64))
            path_img = path + filename
            cv.imwrite(path_img ,img)
            
        except Exception as e:
            continue
            
resize_images(famale_path)
resize_images(male_path)
```

### Get Y label images

```python
def get_images(path):
    label_array = []
    all_images_array = []
    
    for filenames in os.listdir(path):
        try:
            if re.match(r"famales", filenames):
                label_array.append(1)
            else:
                label_array.append(0)
            
            img = cv.imread(path + filenames)
            img_array = np.array(img)
            all_images_array.append(img_array)
            
        except Exception as e:
            continue
    return np.array(all_images_array), np.array(label_array).reshape(1, -1)

famale_img, famale_label = get_images(famale_path)
male_img, male_label = get_images(male_path)
```

```python
print("famale_img: ", famale_img.shape)
print("male_img: ", male_img.shape)

print("famale_label: ", famale_label.shape)
print("male_label: ", male_label.shape)
```

```python
x = np.concatenate((famale_img, male_img), axis = 0)
y = np.concatenate((famale_label.T, male_label.T), axis = 0)
# y = y.T

print("x shape:", x.shape)
print("y shape:", y.shape)
```

### Test/Train Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)

y_train = y_train.T
y_test = y_test.T
```

```python
print("x_train shape: ", x_train.shape)
print("x_test shape: ", x_test.shape)
print("y_train shape: ", y_train.shape)
print("y_test shape: ", y_test.shape)
```

```python
index = 100
classes_list = ["male", "famale"]
classes = np.array(classes_list)
print("y: ", str(y_train.reshape(1, -1)[:,index]), "It is a ", str(np.squeeze(classes[y_train.reshape(1, -1)[:, index]])), " eyes")
plt.imshow(x_train[index])

plt.axis("off")
plt.show()
```

  ### Normalization

```python
import tensorflow as tf
```

```python
x_train_flatten = x_train.reshape(x_train.shape[0], -1).T
x_test_flatten = x_test.reshape(x_test.shape[0], -1).T

print("x_test shape: ", x_test_flatten.shape)
print("x_train shape: ", x_train_flatten.shape)

x_train = x_train_flatten / 255
x_test = x_test_flatten / 255
```
