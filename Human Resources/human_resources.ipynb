```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("HRDataset_v14.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
Salary = data["Salary"]
Termd = data["Termd"]
EngagementSurvey = data["EngagementSurvey"]
EmpSatisfaction = data["EmpSatisfaction"]
SpecialProjectsCount = data["SpecialProjectsCount"]
Absences = data["Absences"]
```

```python
x = pd.concat([Termd, EngagementSurvey], axis = 1)
x = pd.concat([x, EmpSatisfaction], axis = 1)
x = pd.concat([x, SpecialProjectsCount], axis = 1)
x = pd.concat([x, Absences], axis = 1)

y = Salary
```

```python
x.head()
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.40, random_state = 33)
```

```python
X_train
```

```python
y.head()
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train, Y_train)
```

```python
predict = linear_regression.predict(X_test)
```

```python
predict
```

```python
from sklearn.metrics import accuracy_score, r2_score
```

```python
test_data_r2_score = r2_score(Y_test, predict)
```

```python
test_data_r2_score
```
