## Preprocessing

#### Renamed files

```python
def renamed_files(path, label):
    i = 0
    for filenames in os.listdir(path):
        try:
            f, extension = os.path.splitext(path + filenames)
            scr = path + filenames
            dst = path + label + str(i) + extension
            os.rename(scr, dst)
            i += 1
        except Exception as e:
            print(e)
            i += 1

test_apples = "test//apples//"
train_apples = "train//apples//"
test_tomatoes = "test//tomatoes//"
train_tomatoes = "train//tomatoes//"

apples_label = "apples"
tomatoes_label = "tomatoes"

renamed_files(test_apples, apples_label)
renamed_files(train_apples, apples_label)
renamed_files(test_tomatoes, tomatoes_label)
renamed_files(train_tomatoes, tomatoes_label)
```

#### Resized Image

```python
def resized_image(scr_path):
    for filenamed in os.listdir(scr_path):
        try:
            img = Image.open(scr_path + filenamed)
            new_img = img.resize((64,64))
            
            new_img.save(scr_path + filenamed)
            print('Resized and saved {} successfully.'.format(filenamed))
        
        except Exception as ex:
            print(ex)
            continue

scr_test_apples = "test//apples//"
scr_train_apples = "train//apples//"
scr_test_tomatoes = "test//tomatoes//"
scr_train_tomatoes = "train//tomatoes//"

resized_image(scr_test_apples)
resized_image(scr_train_apples)
resized_image(scr_test_tomatoes)
resized_image(scr_train_tomatoes)
```

##### Egitim sürecini yeniden boyutlandır.

```python
def get_data(path):
    all_images_as_array = []
    label = []

    for filename in os.listdir(path):
        try:
            if re.match(r'apples', filename):
                label.append(1)
            else:
                label.append(0)

            img = Image.open(path + filename)
            np_array = np.asarray(img)
            #l, b, c = np_array.shape
            #np_array = np_array.reshape(l*b*c,)
            all_images_as_array.append(np_array)
        except:
            continue
    return np.array(all_images_as_array), np.array(label).reshape(1, -1)


path_to_train_set_apples = "train//apples//"
path_to_test_set_apples = "test//apples//"
path_to_train_set_tomatoes = "train//tomatoes//"
path_to_test_set_tomatoes = "test//tomatoes//"

train_set_x_apples, train_set_y_apples = get_data(path_to_train_set_apples)
test_set_x_apples, test_set_y_apples = get_data(path_to_test_set_apples)
train_set_x_tomatoes, train_set_y_tomatoes = get_data(path_to_train_set_tomatoes)
test_set_x_tomatoes, test_set_y_tomatoes = get_data(path_to_test_set_tomatoes)

print('X_train set apples : ', train_set_x_apples)
print('y_train set apples : ', train_set_y_apples)
print('X_test set apples : ', test_set_x_apples)
print('y_test set apples : ', test_set_y_apples)


print('X_train set tomatoes : ', train_set_x_tomatoes)
print('y_train set tomatoes : ', train_set_y_tomatoes)
print('X_test set tomatoes : ', test_set_x_tomatoes)
print('y_test set tomatoes : ', test_set_y_tomatoes)
```

## 1. Logistic Regression
![The San Juan Mountains are beautiful!](Logistic_reg.jpg "Logistic Regression")

### 1.1 Initialize Paramater 

```python
def initialize_weight_bias(dim):
    w = np.zeros((dim, 1))
    b = 0
    return w, b
```

### 1.2 Sigmoid 

```python
def sigmoid(z):
    y_head = 1 / (1 + np.exp(-z))
    return y_head
```

### 1.3 Implement Forward Backward Propogation, Iteration and Learning Rate

```python
def propogate(w, b, x_train, y_train, number_iteration, learning_rate, print_cost):
    costs = []
    
    for i in range(number_iteration):
        
        # Get number
        m = x_train.shape[1]
    
        # forward propogation
        y_head = sigmoid(np.dot(w.T, x_train) + b)                                              # compute activation
        cost = (-1/m)* np.sum((y_train*np.log(y_head)) + (1 - y_train)*np.log(1 - y_head))      # compute cost
        cost = np.squeeze(cost)


        
        # Backward propogation
            ## ∂J / ∂w = (1 / m) * (x * (yhead − y)T)
            ## ∂J / ∂b = (1 / m) * (m∑(yhead − y))
        derivative_weight = (np.dot(x_train, (y_head - y_train).T)) / m
        derivative_bias = (np.sum(y_head - y_train)) / m
        
        w = w - learning_rate * derivative_weight
        b = b - learning_rate * derivative_bias
        
        params = {"w": w, "b": b}
        grads = {"derivative_weight": derivative_weight, "derivative_bias": derivative_bias}
        
        if i % 10 == 0:
            costs.append(cost)
        
        if print_cost and i % 10 == 0:
            print("Cost after iteration %i: %f" %(i, cost))
            
    return params, grads, costs
```

### 1.4 Implement Predict

```python
def predict(w, b, x):
    m = x.shape[1]
    
    y_prediction = np.zeros((1, m))
    
    w = w.reshape(x.shape[0], 1)
    
    y_head = sigmoid(np.dot(w.T, x) + b)
    
    for i in range(y_head.shape[1]):
        y_prediction[0, i] = 0 if y_head[0, i] < 0.5 else 1
    
    return y_prediction
```

### Logistic Regression

```python
def logistic_regression(x_train, y_train, x_test, y_test, number_iteration, learning_rate, print_cost):
    
    w, b = initialize_weight_bias(x_train.shape[0])
    params, grads, costs = propogate(w, b, x_train, y_train, number_iteration, learning_rate, print_cost)
    
    w = params["w"]
    b = params["b"]
    
    y_prediction_train = predict(w, b, x_train)
    y_prediction_test = predict(w, b, x_test)
    
    # Print train/test Errors
    print("train accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_train - y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_test - y_test)) * 100))
    
    d = {"costs": costs,
         "Y_prediction_test": y_prediction_test, 
         "Y_prediction_train" : y_prediction_train, 
         "w" : w, 
         "b" : b,
         "learning_rate" : learning_rate,
         "num_iterations": number_iteration}
    
    return d
```

```python
d = logistic_regression(x_train, y_train, x_test, y_test, number_iteration = 150, learning_rate = 0.001, print_cost = True)
```

###### Plot 

```python
 # Plot learning curve (with costs)
costs = np.squeeze(d['costs'])
plt.plot(costs)
plt.ylabel('cost')
plt.xlabel('iterations (per hundreds)')
plt.title("Learning rate =" + str(d["learning_rate"]))
plt.show()
```

##### Learning Rates

```python
learning_rates = [0.01, 0.001, 0.0001]
models = {}
for i in learning_rates:
    print ("learning rate is: " + str(i))
    models[str(i)] = logistic_regression(x_train, y_train, x_test, y_test, number_iteration = 150, learning_rate = i, print_cost = False)
    print ('\n' + "-------------------------------------------------------" + '\n')

for i in learning_rates:
    plt.plot(np.squeeze(models[str(i)]["costs"]), label= str(models[str(i)]["learning_rate"]))

plt.ylabel('cost')
plt.xlabel('iterations (hundreds)')

legend = plt.legend(loc='upper center', shadow=True)
frame = legend.get_frame()
frame.set_facecolor('0.90')
plt.show()
```
