```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("classification.csv")
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.head()
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1:]
```

```python
sns.relplot(data = data, x = "age", y = "interest", hue = "success")
```

```python
sns.countplot(x = data["success"], data = data)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, random_state = 0, test_size = 0.33)
```

```python
classification = SVC(kernel ="rbf", degree = 3, gamma = "scale")
```

```python
classification.fit(X_train, Y_train)
```

```python
predict = classification.predict(X_test)
```

```python
from sklearn.metrics import f1_score
```

```python
score = f1_score(Y_test, predict)
```

```python
score
```

```python
input_data = (age, interest)

#Changing Input Data As Numpy Array
input_data_array = np.asarray(input_data)

# Reshape the data as we are predicting for only one instance
input_data_reshape = input_data_array.reshape(1,-1)

# Prediction
prediction = classification.predict(input_data_reshape)
#print(prediction)

if prediction == 0:
    print('Not Successful')
else:
    print('Successful')
```
