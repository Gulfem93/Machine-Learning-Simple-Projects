```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn import svm
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("diabetes2.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
x = data.iloc[:,:-1]
y = data["Outcome"]
```

```python
data.Outcome.value_counts().plot(kind='bar')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
standart_data = sc.fit_transform(x)

x = standart_data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.33, random_state = 33)
```

```python
classifier = svm.SVC(kernel="linear")
classifier.fit(X_train, Y_train)
```

```python
predict = classifier.predict(X_test)
test_data_accurancy_score = accuracy_score(predict, Y_test)
```

```python
test_data_accurancy_score
```

```python
input_data1 = (6,148,72,35,0,33.6,0.627,50)
input_data2 = (1,85,66,29,0,26.6,0.351,31)

#Changing Input Data As Numpy Array
input_data_array = np.asarray(input_data1)

# Reshape the data as we are predicting for only one instance
input_data_reshape = input_data_array.reshape(1,-1)

# Standardize the data
std_data = sc.transform(input_data_reshape)

# Prediction
prediction = classifier.predict(std_data)
#print(prediction)

if prediction == 0:
    print('Person is Not Diabetic')
else:
    print('Person is Diabetic')
```
